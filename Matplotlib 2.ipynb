# Savefig

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line4",dpi=1000,facecolor="g",transparent=False,bbox_inches="tight")

plt.show()
```

# Text

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("wscube",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.xlabel("python",fontsize=15)

plt.text(2,5,"java",fontsize=15,style="italic",bbox={"facecolor":"red"})
plt.annotate("python",xy=(2,1),xytext=(3,3),arrowprops=dict(facecolor="black",shrink=100))

plt.show()
```
