# Bar plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.xlabel("language",fontsize=20)
plt.ylabel("no",fontsize=10)
plt.title("wscube",fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.2,color=c,align="edge",edgecolor="r",linewidth=5,linestyle=":",alpha=0.4)    #align="center"
plt.show()
```

```python
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=20)
plt.ylabel("no",fontsize=10)
plt.title("wscube",fontsize=15)

plt.bar(p,y,width,color="r",label="popularity") 
plt.bar(p1,z,width,color="y",label="popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel("language",fontsize=20)
plt.ylabel("no",fontsize=10)
plt.title("wscube",fontsize=15)

plt.barh(x,y,color="r",label="popularity") 
plt.barh(x,z,color="y",label="popularity1")

plt.legend()
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[400,200,400,300,200,100,600]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.9,marker="*",edgecolor="g",linewidth=2)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=20)
plt.ylabel("no",fontsize=25)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,6,2]
colors=[10,20,30,40,50,60,70]
sizes=[400,200,400,300,200,100,600]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis",alpha=0.8)
plt.scatter(day,no2,c="r",s=sizes,alpha=0.8)
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=20)
plt.ylabel("no",fontsize=25)
plt.show()
```

# Fill_Between

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x,y)
plt.title("python")
plt.xlabel("x.axis")
plt.ylabel("y.axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title("python")
plt.xlabel("x.axis")
plt.ylabel("y.axis")

plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>2)&(x<=4),alpha=0.3)
plt.title("python")
plt.xlabel("x.axis")
plt.ylabel("y.axis")

plt.show()
```
